```python
def data():
        csv_dir =  root_dir +"/DataNew4/CSV/"   
        (dftrain, dfval, dftest) = read_dataframes(csv_dir)
        height_imag = 150 
        width_imag = 150
        x_train  = load_images_tf(dftrain,height_imag,width_imag)
        x_val    = load_images_tf(dfval,height_imag,width_imag)
        x_test   = load_images_tf(dftest,height_imag,width_imag)
        y_train  = create_label_tensor(dftrain)
        y_val    = create_label_tensor(dfval)
        y_test   = create_label_tensor(dftest)
        return x_train, y_train, x_val, y_val, x_test, y_test
  
```

```python
def create_model(x_train, y_train, x_val, y_val, x_test, y_test):
    from hyperopt import Trials, STATUS_OK, tpe
    from keras import layers
    from keras import models
    from hyperas import optim
    from hyperas.distributions import choice, uniform
    import numpy as np
    from keras.utils import np_utils
    
    model = models.Sequential ()
    model.add(layers.Conv2D({{choice([32, 64])}}, (3,3), activation='relu', input_shape=(150,150,3)))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Conv2D({{choice([32, 64, 128])}}, (3,3), activation='relu'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Conv2D({{choice([32, 64, 128])}}, (3,3), activation='relu'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    

# Model Compile

# Compile Network 
    from keras import optimizers 
    model.compile ( loss='binary_crossentropy',
               optimizer = {{choice(['rmsprop', 'adam', 'sgd'])}},
               metrics= ['acc'])
    early_stopping = EarlyStopping(monitor='val_loss', patience=4)
    
    epochs = 100
    
    result = model.fit(x_train, y_train,
                      batch_size={{choice([32, 64, 128])}},
                      epochs=100,
                      verbose=2,
                      validation_data=(x_val, y_val),
                      callbacks=[early_stopping])
    
#get the highest validation accuracy of the training epochs
    validation_acc = np.amax(result.history['val_acc']) 
    print('Best validation acc of epoch:', validation_acc)
    return {'loss': -validation_acc, 'status': STATUS_OK, 'model': model}
```

```python
if __name__ == '__main__':
    from hyperas import optim
    from hyperopt import Trials, STATUS_OK, tpe
    best_run, best_model = optim.minimize(model=create_model,
                                          data=data,
                                          algo=tpe.suggest,
                                          max_evals=5,
                                          trials=Trials(),
                                          eval_space=True,
                                          notebook_name='ModelK13_Hyperas')
    x_train, y_train, x_val, y_val, x_test, y_test = data()
    print("Evalutation of best performing model:")
    print(best_model.evaluate(x_test, y_test))
    print("Best performing model chosen hyper-parameters:")
    print(best_run)
    
```
