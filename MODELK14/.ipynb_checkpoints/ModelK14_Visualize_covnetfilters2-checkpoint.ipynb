```python
# Load model ( We work with a model build with the sequential API )
# We load a model build with sequential model no functional API - ModelK3_1 - Augmentation Model 
Model_name_load = 'ModelK3_1'
model = model_load_tf (model_bin_dir, Model_name_load)
model.summary()
```

```python
#from keras import backend as K
import tensorflow.keras.backend as K
from vis.utils import utils
from vis.visualization import visualize_activation, get_num_filters
from vis.input_modifiers import Jitter

max_filters = 40
selected_indices = []
vis_images = [[], [], []]
i = 0
selected_filters = [[0, 5, 10, 15, 20, 25, 30, 31], 
    [4, 6, 10, 12, 14, 16, 18, 25], 
    [16, 67, 100, 110, 115, 119, 125, 127]]
    

for layer_name in ['conv2d', 'conv2d_1', 'conv2d_2']:
    layer_idx = utils.find_layer_idx(model, layer_name)

    # Visualize all filters in this layer.
    if selected_filters:
        filters = selected_filters[i]
    else:
        filters = sorted(np.random.permutation(get_num_filters(model.layers[layer_idx]))[:max_filters])
    selected_indices.append(filters)

    # Generate input image for each filter.
    for idx in filters:
        img = visualize_activation(model, layer_idx, filter_indices=idx, tv_weight=0., 
                                   input_modifiers=[Jitter(0.05)], max_iter=300) 
        img = utils.draw_text(img, 'Filter {}'.format(idx))  
        vis_images[i].append(img)

    # Generate stitched image palette with 4 cols so we get 2 rows.
    stitched = utils.stitch_images(vis_images[i], cols=4)    
    plt.figure(figsize=(20, 30))
    plt.title(layer_name)
    plt.axis('off')
    plt.imshow(stitched)
    plt.show()
    i += 1
```
