```python
from keras.preprocessing import image
import numpy as np
import pandas as pd 


dftrain = pd.DataFrame()     
dftrain = pd.read_csv(csv_dir+'dftrain.csv')
print(dftrain.shape )


# One image dir 
image_number = 60  # Select the number of image
one_imag_dir  = imag_dir + dftrain['file_name_ext'][image_number]
print ( one_imag_dir)


img = image.load_img (one_imag_dir, target_size=(150,150))
img_tensor = image.img_to_array (img)
img_tensor = np.expand_dims ( img_tensor, axis = 0)
img_tensor /= 255.
print (img_tensor.shape)
```

# IMAGE  = 60

```python
import matplotlib.pyplot as plt
plt.imshow(img_tensor[0])
plt.show()
```

```python
# Load model ( We work with a model build with the sequential API )
# We load a model build with sequential model no functional API - ModelK3_1 - Augmentation Model 
Model_name_load = 'ModelK3_1'
model = model_load_tf (model_bin_dir, Model_name_load)
model.summary()
```

```python
for layer in model.layers[:8]:
    print ( "layer name:", layer.name, layer.trainable)
```

```python
from keras import models 
# Extract the outputs of the top eight layers
layer_outputs = [layer.output for layer in model.layers[:8]]
# Creates a model that will return these outputs, given the model input
activation_model = tf.keras.models.Model(inputs=model.input, outputs=layer_outputs)
activations = activation_model.predict(img_tensor)
```

```python
# First conv layer activation
first_layer_activation = activations[0]
print ( first_layer_activation.shape)
```

```python
# Second Conv layer activation
second_layer_activation = activations[2]
print ( second_layer_activation.shape)
```

```python
# Third Conv layer activation
third_layer_activation = activations[4]
print ( third_layer_activation.shape)
```
